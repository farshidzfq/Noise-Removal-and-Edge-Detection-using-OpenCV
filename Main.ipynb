```python
# نصب کتابخانه OpenCV (در صورت نیاز)
!pip install opencv-python-headless

# وارد کردن کتابخانه‌های لازم
import cv2
import numpy as np
import matplotlib.pyplot as plt

# ایجاد تصویر نمونه با نویز (می‌توانید مسیر تصویر خود را جایگزین کنید)
# دانلود یک تصویر سیاه و سفید با نویز
!wget https://raw.githubusercontent.com/opencv/opencv/master/samples/data/stuff.jpg -O noisy_image.jpg

# خواندن تصویر
image = cv2.imread("noisy_image.jpg", cv2.IMREAD_GRAYSCALE)

# نمایش تصویر اصلی
plt.imshow(image, cmap="gray")
plt.title("Original Noisy Image")
plt.axis("off")
plt.show()

# حذف نویز با استفاده از فیلتر میانگین‌گیری Gaussian
denoised_image = cv2.GaussianBlur(image, (5, 5), 0)

# نمایش تصویر نویززدایی‌شده
plt.imshow(denoised_image, cmap="gray")
plt.title("Denoised Image")
plt.axis("off")
plt.show()

# تشخیص لبه‌ها با استفاده از الگوریتم Canny
edges = cv2.Canny(denoised_image, threshold1=50, threshold2=150)

# نمایش تصویر با لبه‌های استخراج‌شده
plt.imshow(edges, cmap="gray")
plt.title("Edge Detection (Canny)")
plt.axis("off")
plt.show()
```
